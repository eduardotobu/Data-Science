# Whitespace and indentation
If you’ve been working in other programming languages such as Java, C#, or C/C++, you know that these languages use semicolons (;) to separate the statements.

However, Python uses whitespace and indentation to construct the code structure.

The following shows a snippet of Python code:

```python
# define main function to print out something
def main():
    i = 1 #first level of identation
    max = 10
    while (i < max):
        print(i) #second level of identation
        i = i + 1

# call function main 
main()
```

# Strings
A string is a series of characters. In Python, anything inside quotes is a string. And you can use either single or double quotes. For example:

```python
message = 'This is a string in Python'
print(message)
message = "This is also a string"
print(message)
```

To escape the quotes, you use the backslash ( \\ ). For example:

```python
message = 'It\'s also a valid string' #backlash for escaping the single quote
print(message)
```

The Python interpreter will treat the backslash character (\) special. If you don’t want it to do so, you can use raw strings by adding the letter r before the first quote. For example:

```python
message = r'C:\python\bin' #raw string
print(message)
```

# Difference between 'and' operator and '&' operator
'AND' operator evaluates first the value from the left and then the value from the right, if both values are true, the result will be the value from the right.
'&' operator evaluates both values on a bitwise sense. <br>
i.e. 4(in binary 0100) & 14(in binary 1110) = 0100 (and applies to every bit)

```python
a = 14
b = 4

print(b and a)  # print_stat1
print(b & a)  # print_stat2
```

# Difference between 'and' operator and '&' operator
'OR' operator evaluates first the value from the left and then the value from the right, if both values are true, the result will be the value from the left.
'|' operator evaluates both values on a bitwise sense. <br>
i.e. 4(in binary 0100) ! 14(in binary 1110) = 1110 (and applies to every bit).

```python
a = 14
b = 4

print(b or a)  # print_stat1
print(b | a)
```

# Alternative form of elif statement (Ternary operator)

In this statement, the left side of the assignment operator (=) is the variable ticket_price.

The expression on the right side returns 20 if the age is greater than or equal to 18 or 5 otherwise.

The following syntax is called a ternary operator in Python:

```python
age = input('Enter your age:')

#Ternary operator.
ticket_price = 20 if int(age) >= 18 else 5

print(f"The ticket price is {ticket_price}")
```

# Two uses of a while condition

```python
#1) Simple while statement
maxi = 5
counter = 0

while counter < maxi:
    print(counter)
    counter += 1
```

```python
#2) Using the Python while statement to build a simple command prompt program
command = ''

while command.lower() != 'quit':
    #ask the user for an input
    command = input('>')
    print(f"Echo: {command}")
```

# Use cases of break statement

When you use the break statement in a nested loop, it’ll terminate the innermost loop. For example:

This example uses two for loops to show the coordinates from (0,0) to (5,5) on the screen.
The break statement in the nested loop terminates the innermost loop when the y is greater than one.
Therefore, you only see the coordinates whose y values are zero and one.

```python
for x in range(5):
    for y in range(5):
        # terminate the innermost loop
        if y > 1:
            break
        # show coordinates on the screen
        print(f"({x},{y})")
```

# Using Python break statement with a while loop
The following example uses the break statement inside a while loop.

It’ll prompt you for entering your favorite color. The program will stop once you enter quit:

```python

print('-- Help: type quit to exit --')
#an indefinite loop
while True:
    #asking for an input
    color = input('Enter your favorite color:')
    if color.lower() == 'quit':
        #escape from the loop
        break
```
