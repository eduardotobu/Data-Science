# Productos ABC XYZ

## Venta diaria
Obtenemos la venta diaria de las últimas 8 semanas, sin contar la semana en curso, filtrado por catálogo al día de hoy, además obtenemos las métricas necesarias para calcular el ABC (total_revenue) y el XYZ (Coeficiente de Variación).

```python
%%time

#corremos el query para obtener la venta diaria

query = '''
WITH
	combos AS
	 (
		WITH pre_combos AS (
		    SELECT
		        parent_product_id,
		        COUNT(child_product_id) AS conteo
		    FROM product_combos
		    WHERE child_product_id != 5166
		    GROUP BY 1
		    HAVING conteo > 1
			),
			combos AS (
			    SELECT DISTINCT
			        parent_product_id
			    FROM product_combos pc
			    WHERE parent_product_id IN(
			            SELECT
			                parent_product_id
			            FROM pre_combos)
			            )
		SELECT * 
		FROM combos
		)
		
	,adicionales AS (
		SELECT 
		    pg.nombre, 
		    bp.name,
		    bp.codigo
		  FROM productos_grupo pg
		  JOIN productos_grupo_detalles pgd on pgd.productos_grupo_id=pg.id
		  JOIN bi_products bp on bp.id = pgd.producto_id
			)
	
	,product_sales AS
		(
		SELECT
		os.delivery_date AS date
			,CASE WHEN (os.workshop_id = 1966 OR os.workshop_id = 396) THEN 103455 ELSE os.workshop_id END AS workshop_id
			,CASE WHEN (os.workshop_name = 'CDMX' OR os.workshop_name = 'EDOMEX') THEN 'Valle de Mex' ELSE os.workshop_name END AS workshop_name
			,os.product_category_name AS product_category
		    ,os.product_sku
		    ,os.product_id
		    ,os.product_name
		    ,SUM(os.product_quantity) AS units
		    ,SUM(CASE
		        WHEN os.product_id in (10309, 10310) THEN (os.product_sale_share - os.product_delivery_share  + os.product_discount_share)   
		        WHEN os.product_id in (12377) THEN (os.product_sale_share - os.product_delivery_share  + os.product_discount_share)/(1+0.0586) 
		        WHEN os.product_category_id IN (7,13) THEN (os.product_sale_share - os.product_delivery_share  + os.product_discount_share) / (1+ 0.048)
		        WHEN os.product_category_id IN (9) THEN (os.product_sale_share - os.product_delivery_share  + os.product_discount_share) / (1+ 0.08)
		        ELSE (os.product_sale_share - os.product_delivery_share  + os.product_discount_share) / (1+0.16) END) AS revenue
		    
		FROM order_sales os		
		WHERE TRUE
			AND os.order_status_id = 3
			AND os.is_test_order = FALSE
			AND os.is_sales_valid = TRUE
			AND os.product_sku NOT IN (SELECT DISTINCT codigo FROM adicionales)
			AND os.product_id NOT IN (SELECT DISTINCT parent_product_id FROM combos)
			AND os.delivery_date BETWEEN (date_trunc('week',CURRENT_DATE) - interval'12 weeks') AND (date_trunc('week', CURRENT_DATE) - interval'1 days')
			AND os.workshop_id IS NOT NULL
            AND os.delivery_date NOT BETWEEN '2022-08-27' AND '2022-08-29'
		GROUP BY 1,2,3,4,5,6,7
		ORDER BY 2,5,1
		)
		
	,dates AS
		(
		SELECT 
			DISTINCT date_updated,
			workshop_id,
			product_id
		FROM product_inactivations
		WHERE TRUE
			AND date_updated BETWEEN (date_trunc('week',CURRENT_DATE) - interval'8 weeks') AND (date_trunc('week', CURRENT_DATE) - interval'1 days')
		)
SELECT
	d.date_updated AS date,
	pro.workshop_id, 
	pro.workshop_name, 
	pro.product_category,
	pro.product_sku, 
	pro.product_id, 
	pro.product_name,
	COALESCE(ps.units,0) AS units,
	COALESCE(ps.revenue,0) AS revenue
	
FROM dates d
	JOIN (
	SELECT DISTINCT workshop_id, workshop_name, product_category, product_sku, product_id, product_name 
	FROM product_sales) pro ON d.workshop_id = pro.workshop_id AND d.product_id = pro.product_id
	LEFT JOIN product_sales ps ON d.date_updated = ps.date AND pro.workshop_id = ps.workshop_id AND pro.product_id = ps.product_id

WHERE TRUE
    AND pro.product_id IN (SELECT id FROM productos WHERE activo = 'Si' AND discontinued = 'no')
    
ORDER BY 5,2,1
'''

#limpiamos el query
query = query.replace('\n',' ').replace('\t',' ')

#metemos la información consultada del query a nuestro dataframe
df_venta_diaria = pd.read_sql(query, engine)

df_venta_diaria = df_venta_diaria.fillna(0)

#Para las entradas de productos que no tienen categoría.
df_venta_diaria = df_venta_diaria.loc[df_venta_diaria['product_category'] != 0]

df_venta_diaria
```

## Clasificación XYZ.
Clasificamos los materiales según la variabilidad en la demanda, siendo X los materiales con menos variabilidad en el consumo y Z los materiales con más variabilidad.

```python
df_xyz_data = df_venta_diaria
df_xyz_data = df_xyz_data.groupby(['workshop_id','workshop_name','product_category','product_sku','product_id','product_name']).agg(total_product_units = ('units','sum'),
                                                                                                                               avg_units = ('units','mean'), 
                                                                                                                                std_units = ('units','std'),
                                                                                                                                total_revenue = ('revenue','sum')).reset_index()
#calculo de coeficiente de variacion.
df_xyz_data['CV'] = df_xyz_data['std_units']/df_xyz_data['avg_units']

#clasificacion XYZ
df_xyz_data['xyz'] = 'Z'
df_xyz_data.loc[df_xyz_data['CV'] < 0.95, 'xyz'] = 'Y'
df_xyz_data.loc[df_xyz_data['CV'] < 0.45, 'xyz'] = 'X'

df_xyz_data
```

## Clasificación ABC
Clasificamos los productos como ABC por workshop-categoría dependiendo del revenue total del producto. </br>
Los productos que representan el primer 80% del revenue por taller-categoría son clasificados como A, el siguiente 15% como B y el último 5% como C.

```python
df_product_abc = pd.DataFrame(columns = ['workshop_id', 'workshop_name','product_category','product_sku','product_id','product_name', 'total_revenue','running_percent', 'abc','xyz'])

for workshop_id in list(df_xyz_data['workshop_id'].unique()):
    df_vd_workshop = df_xyz_data.loc[df_xyz_data['workshop_id'] == workshop_id]
   
    for product_category in list(df_vd_workshop['product_category'].unique()):
        df_vd_category = df_vd_workshop.loc[df_vd_workshop['product_category'] == product_category]
        df_vd_category = df_vd_category.sort_values(by=['total_revenue'], ascending=False)
        df_vd_category['rank'] = df_vd_category['total_revenue'].rank(ascending = False)
        
        df_vd_category['running_percent'] = (df_vd_category['total_revenue'].cumsum()/df_vd_category['total_revenue'].sum()) 
        df_vd_category['abc'] = ''
        
        df_vd_category['abc'][0: int(0.10 * df_vd_category['rank'].max())] = 'A'
        
        df_vd_category.loc[df_vd_category['running_percent'] <= .80, 'abc'] = 'A'
        df_vd_category.loc[(df_vd_category['running_percent'] > .80) & (df_vd_category['running_percent'] <= .95), 'abc'] = 'B'
        df_vd_category.loc[df_vd_category['running_percent'] > .95, 'abc'] = 'C' 
        
        #para las categorías que tienen 3 productos o menos.
        df_vd_category['abc'].iloc[0] = 'A'
    
        if df_vd_category['product_id'].count() == 2:
            df_vd_category['abc'].iloc[0] = 'A'
            df_vd_category['abc'].iloc[1] = 'B'
            
        elif df_vd_category['product_id'].count() == 3:
            df_vd_category['abc'].iloc[0] = 'A'
            df_vd_category['abc'].iloc[1] = 'B'
            df_vd_category['abc'].iloc[2] = 'C'
            
        try:
            if df_vd_category['running_percent'].loc[df_vd_category['abc'] == 'B'].iloc[0] < .85:
                df_vd_category['abc'].loc[df_vd_category['running_percent'] == df_vd_category['running_percent'].loc[df_vd_category['abc'] == 'B'].iloc[0]] = 'A'
        except:
            pass
        #ponemos en c los que tengan revenue 0.
        df_vd_category.loc[df_vd_category['total_revenue'] == 0, 'abc'] = 'C'
        
        df_vd_category = df_vd_category[['workshop_id','workshop_name', 'product_category','product_sku','product_id','product_name','total_revenue','running_percent','abc','xyz']]
        df_product_abc = df_product_abc.append(df_vd_category)
        
df_product_abc = df_product_abc.fillna(0)
df_product_abc = df_product_abc.reset_index(drop = True)

#lista_productos_estrella = list(df_starp_w_p['workshop_id'].astype(str) + df_starp_w_p['product_id'].astype(str))
#df_product_abc.loc[(df_product_abc['workshop_id'].astype(str) + df_product_abc['product_id'].astype(str)).isin(lista_productos_estrella), 'abc'] = 'A'

df_product_abc
```

# Materiales ABC XYZ

## Explosión de materiales.
Agregamos recetario para explotar la lista de productos y agrupamos los consumos por material.

```python
%%time
#corremos el query para obtener el recetario.

query = '''
SELECT 
	prd.id AS product_id,
	mcat.nombre AS material_category,
	m.id AS material_id,
	m.nombre AS material_name,
	CASE WHEN pmd.material_id IN (1550,1551,112,1457,17,16,2619,2727,3192,3221) THEN 1 ELSE pmd.cantidad END AS material_quantity
	
FROM productos prd
	LEFT JOIN pmaterialdetalles pmd ON pmd.producto_id = COALESCE(prd.related_product_id, prd.id)
	LEFT JOIN materiales m ON pmd.material_id = m.id
	LEFT JOIN mcategorias mcat ON m.mcategoria_id = mcat.id
	
WHERE TRUE
	AND prd.activo = 'Si'
	AND prd.discontinued = 'no'

ORDER BY 1,5
'''

#limpiamos el query
query = query.replace('\n',' ').replace('\t',' ')

#metemos la información consultada del query a nuestro dataframe
df_recetario = pd.read_sql(query, engine)
df_recetario = df_recetario.fillna(0)
df_recetario = df_recetario.astype({'material_id':'int'})
df_consumo_material = df_venta_diaria.merge(df_recetario, how = 'left', on ='product_id')
df_consumo_material['material_units'] = df_consumo_material['material_quantity'] * df_consumo_material['units']
df_consumo_material
```

## Agrupar la info por material.

```python
df_grouped = df_consumo_material.copy()

#se agrupa primero por producto, para obtener info diaria de material y obtener el promedio y la desviacion estandar diaria.
df_grouped = df_grouped.groupby(['date','workshop_id','material_id']).agg(total_units = ('material_units','sum')).reset_index()

#obtenemos promedio y desviación estándar.
df_grouped = df_grouped.groupby(['workshop_id','material_id']).agg(total_units = ('total_units','sum'),
                                                                    avg_units = ('total_units','mean'),
                                                                    stdev_units = ('total_units','std'),).reset_index()
df_grouped
```

## Clasificación XYZ

```python
df_material_xyz = df_grouped.copy()

#calculo de coeficiente de variacion
df_material_xyz['CV'] = df_material_xyz['stdev_units'] / df_material_xyz['avg_units']

#clasificacion xyz
df_material_xyz['xyz'] = 'Z'
df_material_xyz.loc[df_material_xyz['CV'] < 0.95, 'xyz'] = 'Y'
df_material_xyz.loc[df_material_xyz['CV'] < 0.45, 'xyz'] = 'X'

df_material_xyz = df_material_xyz[['workshop_id', 'material_id', 'xyz']]
df_material_xyz
```

### Función para conectarse con MFA

```python
df_starp_w_p.to_csv('data/star_products.csv', index = False)

from botocore.exceptions import ClientError, ParamValidationError
from datetime import date
import boto3


def boto3_mfa(device_arn, service):
    """
    Function for generating an aws service client using a MFA device and code.
    Example:
    running:
        s3_client = boto3_mfa('arn:aws:iam::714714037319:mfa/eduardotobu', 's3')
    should prompt the user for a mfa code, afterwards the function returns a 
    boto client for the aws s3 service.
    """
    done = False
    cc = 0
    while not done:
        cc += 1
        mfa_otp = input("Enter the MFA code: ")
        sts = boto3.client('sts')
        try:
            creds = sts.get_session_token(
                DurationSeconds = 900, # 15 min de activo
                SerialNumber = device_arn, 
                TokenCode = mfa_otp
                )
            client = boto3.client(service_name = service,
                        region_name ='us-west-2',
                        aws_access_key_id = creds['Credentials']['AccessKeyId'],
                        aws_secret_access_key = creds['Credentials']['SecretAccessKey'],
                        aws_session_token = creds['Credentials']['SessionToken'])
            done = True
        except ClientError as e:
            print(e)
            done = False # si quieren hacer error handling por aqui se puede
        except ParamValidationError as e:
            print('Token is wrong length')
            done = False
        if cc >= 15:
            print('Too many tries')
            done = True
    return client
```
