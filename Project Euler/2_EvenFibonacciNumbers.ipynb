# Problem 2: Even Fibonacci Numbers.

```python
%%time
def fibonacci_nums(fib_max):
    '''Return sum of even Fibonacci numbers less than the input value.'''
    fib_list = [1,2]
    next_fib = sum(fib_list)
    
    while next_fib < fib_max:
        fib_list.append(next_fib)
        next_fib = fib_list[len(fib_list) - 1] + fib_list[len(fib_list) - 2]
    
    fib_list = [i for i in fib_list if i%2 == 0]
    return sum(fib_list)

fibonacci_nums(4000000)
```
