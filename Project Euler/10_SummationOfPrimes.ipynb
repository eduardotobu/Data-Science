# Problem 10: Summation Of Primes.

```python
%%time
def summation_of_primes(number_below):
    
    '''Returns the sum of all the primes bellow a certain quantity'''
    
    sum_primes = 2

    for n in range(3, number_below):
        for m in range(2,n):
            if n % m == 0:
                break

            if m == n-1:
                sum_primes = n + sum_primes
    
    return sum_primes
    
summation_of_primes(2000000)
```
