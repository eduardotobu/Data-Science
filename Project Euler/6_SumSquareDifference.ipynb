# Problem 6: Sum Square Difference.

```python
%%time
def sum_square_difference(bot, top): 

    '''Return the difference between the sum of the squares and the square of the sum of the numbers in a range.'''   
    
    numbers = range(bot, top + 1)
    square_of_sum = sum(numbers)**2
    sum_of_squares = sum([n**2 for n in numbers])
    ss_difference = square_of_sum - sum_of_squares
    
    return ss_difference

sum_square_difference(1,100)
```
