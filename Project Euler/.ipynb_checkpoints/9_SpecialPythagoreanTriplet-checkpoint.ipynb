# Problem 9: Special Pythagorean Triplet.

```python
%%time
def special_pythagorean_triplet():
    
    '''Returns the product of the exactly pythagorean triplet for wich a + b + c = 1000 '''
    
    for c in range (1000,2,-1):
        for b in range (c,2,-1):
            for a in range(b,2,-1):
                if a + b + c == 1000 and a**2 + b**2 == c**2:
                    
                    return a*b*c
                
special_pythagorean_triplet()
```
