# Problem 7: 10001st Prime.

## My solution:

```python
%%time
def n_prime(n_pr):
    
    '''Return the Nst prime number'''
    
    n_p = 1 #Already set to 1st prime number.
    n = 2 #First prime number.

    if n_pr == 1:
        prime_number = n
    else:
        while n_p < n_pr:
            n += 1 
            for i in range(2, n + 1):
                if n % i == 0:
                    break
                if i == n - 1:
                    prime_number = n
                    n_p += 1
                    
    return prime_number

n_prime(10001)
```

## A fastest approach found on the internet:

```python
%%time
def isPrime(n):
    if n < 2: return "Neither prime, nor composite"
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# returns the nth prime number
def nthPrime(n):
    numberOfPrimes = 0
    prime = 1

    while numberOfPrimes < n:
        prime += 1
        if isPrime(prime):
            numberOfPrimes += 1
    return prime

print(nthPrime(10001))
```
