# Problem 1: Multiples of 3 or 5.

```python
def multiples_3_5(max_n):
    '''Get input from user and calculate sum of multiples of 3 or 5 between 0 and the input.'''
    n_sum = 0
    for n in range(max_n):  
        if n % 3 == 0 or n % 5 == 0:
            n_sum += n
            
    return n_sum
    
multiples_3_5(1000)
```
